### Importação dos dados



```python
#importação das bibliotecas pandas e matplotlib para análise de dados e cálculos
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

#definição das variáveis
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
loja.info()
```

```python
# PASSO 1: Nesta primeira análise, você deve calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento.

#EXPLICAÇÃO:
#criamos uma variável (pois usamos o =) e chamamos as variáveis criadas lá em cima
#.preço pois queremos pegar apenas a coluna preço
#.sum() pois queremos somar a coluna
#.round(2) para arredondar a colua até 2 casas decimais

Total_Faturamento_loja_1 = loja.Preço.sum().round(2)
Total_Faturamento_loja_2 = loja2.Preço.sum().round(2)
Total_Faturamento_loja_3 = loja3.Preço.sum().round(2)
Total_Faturamento_loja_4 = loja4.Preço.sum().round(2)
```

```python
#temos que fazer a impressão agora dessas variáveis
#usamos o f-sting
print(f"O faturamento da loje 1 é de: ${Total_Faturamento_loja_1}")
print(f"O faturamento da loje 2 é de: ${Total_Faturamento_loja_2}")
print(f"O faturamento da loje 3 é de: ${Total_Faturamento_loja_3}")
print(f"O faturamento da loje 4 é de: ${Total_Faturamento_loja_4}")
```

# 2. Vendas por Categoria

```python
#PASSO 2: Neste passo, deve calcular a quantidade de produtos vendidos por categoria em cada loja.
#A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

# EXPLICAÇÃO:
# .groupby('Categoria do Produto'): Agrupa todas as linhas que possuem a mesma categoria.
# [['Produto']]: Selecionamos apenas a coluna 'Produto' para trabalhar após o agrupamento.
# .count(): Conta quantas vezes cada produto apareceu em cada categoria (gera a quantidade vendida).
# .sort_values('Produto', ascending=False): Ordena o resultado do maior para o menor (do mais vendido para o menos vendido).

Venda_Categoria_loja_1 = loja.groupby('Categoria do Produto')[['Produto']].count().sort_values('Produto', ascending= False)
Venda_Categoria_loja_2 = loja2.groupby('Categoria do Produto')[['Produto']].count().sort_values('Produto', ascending= False)
Venda_Categoria_loja_3 = loja3.groupby('Categoria do Produto')[['Produto']].count().sort_values('Produto', ascending= False)
Venda_Categoria_loja_4 = loja4.groupby('Categoria do Produto')[['Produto']].count().sort_values('Produto', ascending= False)
```

```python
print(f'Vendas por categoria da loja 1 {Venda_Categoria_loja_1}')
print(f'Vendas por categoria da loja 2 {Venda_Categoria_loja_2}')
print(f'Vendas por categoria da loja 3 {Venda_Categoria_loja_3}')
print(f'Vendas por categoria da loja 4 {Venda_Categoria_loja_4}')
```

# 3. Média de Avaliação das Lojas


```python
#PASSO 3: Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

def avaliacao_media(loja, numero):
    media = loja['Avaliação da compra'].mean()
    print(f"Avaliação média na Loja {numero}: {media:.2f}")

# Re-load 'loja' DataFrame as it was overwritten by a string variable
# The 'url' variable holds the URL for 'loja_1.csv'
loja = pd.read_csv(url)

# Mostrar resultados para cada loja
avaliacao_media(loja, 1)
avaliacao_media(loja2, 2)
avaliacao_media(loja3, 3)
avaliacao_media(loja4, 4)
```

# 4. Produtos Mais e Menos Vendidos

```python
#PASSO 4: Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.
#Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

urls = [url, url2, url3, url4]

# Ler todos os CSVs e adicionar uma coluna 'Loja'
df_total = pd.concat(
    [pd.read_csv(url).assign(Loja=f"Loja {i+1}") for i, url in enumerate(urls)],
    ignore_index=True
)

# Garantir os tipos corretos
df_total['Preço'] = df_total['Preço'].astype(float)
df_total['Quantidade de parcelas'] = df_total['Quantidade de parcelas'].astype(int)

# A receita é simplesmente o Preço (não multiplicamos pela quantidade de parcelas)
df_total['Receita'] = df_total['Preço']

# Agrupar por Loja e Produto somando a receita
vendas = df_total.groupby(['Loja', 'Produto'])['Receita'].sum().reset_index()

# Função para mostrar os 3 produtos que mais e menos dinheiro geraram
def top_produtos(loja):
    dados = vendas[vendas['Loja'] == loja]

    top_mais_vendidos = dados.sort_values(by='Receita', ascending=False).head(3)
    top_menos_vendidos = dados.sort_values(by='Receita', ascending=True).head(3)

    print(f"\n{loja} - TOP 3 produtos que MAIS dinheiro geraram:")
    for _, linha in top_mais_vendidos.iterrows():
        print(f"  {linha['Produto']} - Receita: R${linha['Receita']:.2f}")

    print(f"\n{loja} - TOP 3 produtos que MENOS dinheiro geraram:")
    for _, linha in top_menos_vendidos.iterrows():
        print(f"  {linha['Produto']} - Receita: R${linha['Receita']:.2f}")

# Lista única de lojas
lojas_unicas = vendas['Loja'].unique()

# Mostrar resultados para cada loja
for loja in lojas_unicas:
    top_produtos(loja)
```

# 5. Frete Médio por Loja

```python
#PASSO 5: Neste passo, vamos calcular o custo médio de frete para cada loja.
#O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas. Sabendo que o cliente que paga o frete.

# Garantir que o custo de envio é numérico
df_total['Frete'] = df_total['Frete'].astype(float)

# Agrupar por loja e calcular a média do custo de envio
envio_medio = df_total.groupby('Loja')['Frete'].mean()

# Mostrar resultados
print("\nEnvio médio por loja:")
for loja, medio in envio_medio.items():
    print(f"{loja}: R${medio:.2f}")
```

# PREPARAÇÃO DOS GRÁFICOS

```python
# Adiciona a coluna 'Tienda_ID' com a etiqueta da loja
loja['Loja_ID'] = 'Loja 1'
loja2['Loja_ID'] = 'Loja 2'
loja3['Loja_ID'] = 'Loja 3'
loja4['Loja_ID'] = 'Loja 4'

# Concatena os DataFrames de todas as lojas em um único DataFrame (df_total)
df_total = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)
```

```python
# Converte a coluna 'Fecha de Compra' para o formato de data e hora, assumindo que o dia vem primeiro (dia/mês/ano).
df_total['Data de Compra'] = pd.to_datetime(df_total['Data da Compra'], dayfirst=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Agrupar dados (presume-se que df_total e ingresos_totales já foram calculados)
ingresos_totales = df_total.groupby('Loja')['Preço'].sum()

# 2. Definimos as cores: Vermelho para a Tienda 4 (a de pior desempenho)
colores = ['#1f77b4', '#1f77b4', '#1f77b4', 'red'] # Azul, Azul, Azul, Vermelho

# Configurando o Grafico
plt.figure(figsize=(8,5))
# Usamos o parâmetro 'color' para aplicar a paleta
ingresos_totales.plot(kind='bar', color=colores)
#Especificação do grafico
plt.title('Receita Total por Loja (Loja 4 = Menor Receita)', fontsize=15)
plt.ylabel('Receita Total (Real)')
plt.xlabel('Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0) # Mantemos o rótulo da loja horizontal
plt.tight_layout()
plt.show()
```

```python
# 2. Linha: Receita Média por Venda
ingreso_promedio = df_total.groupby('Loja')['Preço'].mean()
plt.figure(figsize=(8,5))
ingreso_promedio.plot(kind='bar', color=colores)
plt.title('Receita Média de Venda por Loja')
plt.ylabel('Receita Média (Real)')
plt.xlabel('Loja')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Boxplot: Comparamos a distribuição das Classificações
sns.boxplot(data=df_total, x='Avaliação da compra', y='Frete', palette='pastel')

plt.title('Custo de Envio vs. Classificação do Cliente (Todas as Lojas)', fontsize=15)
plt.xlabel('Classificação do Cliente (1 a 5)')
plt.ylabel('Custo de Envio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# 1. Garantimos que a coluna 'Data de Compra' seja datetime (crucial!)
df_total['Data de Compra'] = pd.to_datetime(df_total['Data da Compra'], dayfirst=True)

# 2. Agrupamos por Mês e por Loja, somando o Lucro Hipotético (usando 'Receita' como proxy)
lucro_mensal = df_total.groupby([
    df_total['Data de Compra'].dt.to_period('M'), # Agrupamento por mês
    'Loja'
])['Receita'].sum().reset_index()

# 3. Convertemos o período para string para plotar
lucro_mensal['Data de Compra'] = lucro_mensal['Data de Compra'].astype(str)

# 4. Plotamos a série temporal
plt.figure(figsize=(12, 6))
# Usamos 'hue='Loja_ID'' para que cada loja seja uma linha diferente
sns.lineplot(data=lucro_mensal, x='Data de Compra', y='Receita', hue='Loja', marker='o')
plt.title('Tendência de faturamento Mensal por Loja', fontsize=15)
plt.ylabel('Faturamento Total Mensal (Real)')
plt.xlabel('Mês')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. Presume-se que lucro_mensal já foi calculado e preparado (from previous cell)
# 2. Definimos a paleta de cores para destacar a Loja 4
color_map = {'Loja 1': '#1f77b4', 'Loja 2': '#ff7f0e', 'Loja 3': '#2ca02c', 'Loja 4': 'red'}

plt.figure(figsize=(12, 5))

for loja_id in lucro_mensal['Loja'].unique():
    data_loja = lucro_mensal[lucro_mensal['Loja'] == loja_id].copy()
    # Ensure 'Data de Compra' is sorted for rolling mean calculation
    data_loja = data_loja.sort_values('Data de Compra')
    data_loja['Media_Movil_6M'] = data_loja['Receita'].rolling(window=12).mean() # Changed 'Lucro_Hipotetico' to 'Receita'

    # Plotamos, usando a cor específica para a loja
    sns.lineplot(data=data_loja, x='Data de Compra', y='Media_Movil_6M',
                 label=loja_id, marker='o', color=color_map[loja_id])

plt.title('Tendência Suavizada do Faturamento Mensal (Média Móvel 6M) - Destacando Loja 4', fontsize=15)
plt.ylabel('Faturamento Mensal Médio Suavizado (Real)') # Updated ylabel
plt.xlabel('Mês')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Lojas')
plt.show()
```

```python
# 1. Presume-se que lucro_mensal já foi calculado e preparado (from previous cell)
# 2. Definimos a paleta de cores para destacar a Loja 4
color_map = {'Loja 1': '#1f77b4', 'Loja 2': '#ff7f0e', 'Loja 3': '#2ca02c', 'Loja 4': 'red'}

plt.figure(figsize=(12, 6))

# Iterate through unique 'Loja_ID' instead of 'Tienda'
for loja_id in lucro_mensal['Loja'].unique():
    data_loja = lucro_mensal[lucro_mensal['Loja'] == loja_id].copy()
    # Ensure 'Data de Compra' is sorted for cumulative sum
    data_loja = data_loja.sort_values('Data de Compra')
    # Calculate accumulated profit instead of Ingresos_Mensuales
    data_loja['Lucro_Acumulado'] = data_loja['Receita'].cumsum() # Changed 'Lucro_Hipotetico' to 'Receita'

    # Plotamos, usando a cor específica para a loja
    sns.lineplot(data=data_loja, x='Data de Compra', y='Lucro_Acumulado',
                 label=loja_id, marker='', color=color_map[loja_id])

plt.title('Faturamento Acumulado Total por Loja - Destacando Loja 4', fontsize=15)
plt.ylabel('Faturamento Acumulado Total (Real)')
plt.xlabel('Mês')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Loja')
plt.tight_layout()
plt.show()
```
